```python
from xgboost import XGBClassifier


from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import numpy as np


# 划分标记数据为训练集和测试集
train_data, test_data1, train_labels, test_labels1 = train_test_split(
    labeled_data.iloc[:, :24], labeled_data.iloc[:, -1], test_size=0.2, random_state=42)


# 初始化提升树分类器
model = XGBClassifier()


# 训练第一个监督模型
model.fit(train_data, train_labels)

print(unlabeled_data)
# 预测未标记数据的类别
unlabeled_predictions = model.predict(unlabeled_data.iloc[:, :24])
unlabeled_probability = model.predict_proba(unlabeled_data.iloc[:, :24])


# 选择满足预定义标准的观测结果，可自己更改概率值
threshold = 0.9  # 预测概率大于90%

# 将预测结果与原始数据结合起来
unlabeled_data['landcover'] = unlabeled_predictions
unlabeled_data['probability'] = unlabeled_probability.max(axis=1)

# 选择满足预定义标准的观测结果
new_labeled_data = unlabeled_data[unlabeled_data['probability'] > threshold].iloc[:, :-1]
unlabeled_data = unlabeled_data[unlabeled_data['probability'] <= threshold].iloc[:, :-2]

# 将这些观测结果添加到标记数据中
labeled_data = pd.concat([labeled_data, new_labeled_data], axis=0)
# 迭代过程
max_iterations = 5 # 指定最大迭代次数,可自己更改，增加迭代次数

for iteration in range(1, max_iterations + 1):
    # 划分新的标记数据为训练集和测试集
    train_data, test_data, train_labels, test_labels = train_test_split(
        labeled_data.iloc[:, :24], labeled_data.iloc[:, -1], test_size=0.2, random_state=42
    )

    # 训练新的监督模型
    model.fit(train_data, train_labels)


    # 预测未标记数据的类别
    unlabeled_predictions = model.predict(unlabeled_data.iloc[:, :24])
    unlabeled_probability = model.predict_proba(unlabeled_data.iloc[:, :24])

    # 将预测结果与原始数据结合起来
    unlabeled_data['landcover'] = unlabeled_predictions
    unlabeled_data['probability'] = unlabeled_probability.max(axis=1)

    # 选择满足预定义标准的观测结果
    new_labeled_data = unlabeled_data[unlabeled_data['probability'] > threshold].iloc[:, :-1]
    unlabeled_data = unlabeled_data[unlabeled_data['probability'] <= threshold].iloc[:, :-2]

    # 将这些观测结果添加到标记数据中
    labeled_data = pd.concat([labeled_data, new_labeled_data], axis=0)

    # 输出当前迭代的准确率
    train_accuracy = accuracy_score(train_labels, model.predict(train_data))
    test_accuracy = accuracy_score(test_labels, model.predict(test_data))
    print(f"Iteration {iteration}: labeled data = {labeled_data.shape[0]}, New labeled data = {new_labeled_data.shape[0]}, Unlabeled data = {unlabeled_data.shape[0]},\
           Train Accuracy = {train_accuracy:.4f}, Test Accuracy = {test_accuracy:.4f}")


# 最终模型在测试集上的准确率
final_test_accuracy = accuracy_score(test_labels1, model.predict(test_data1))

print(f"Final Test Accuracy = {final_test_accuracy:.4f}")
print("Classification report: \n",classification_report(model.predict(test_data),test_labels))
from sklearn.metrics import cohen_kappa_score

kappa = cohen_kappa_score(model.predict(test_data1), test_labels1)
print("Kappa值: ", kappa)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix, cohen_kappa_score

# 计算混淆矩阵
conf_matrix = confusion_matrix(model.predict(test_data),test_labels)

# 计算每个类别的Kappa值
kappa_per_class = []
for i in range(conf_matrix.shape[0]):
    true_positive = conf_matrix[i, i]
    false_positive = conf_matrix[:, i].sum() - true_positive
    false_negative = conf_matrix[i, :].sum() - true_positive
    true_negative = conf_matrix.sum() - true_positive - false_positive - false_negative

    observed_agreement = (true_positive + true_negative) / conf_matrix.sum()
    chance_agreement = ((true_positive + false_positive) * (true_positive + false_negative) +
                        (false_negative + true_negative) * (false_positive + true_negative)) / (conf_matrix.sum() ** 2)

    kappa = (observed_agreement - chance_agreement) / (1 - chance_agreement)
    kappa_per_class.append(kappa)

print("Kappa per class:")
for i, kappa in enumerate(kappa_per_class):
    print(f"Class {i}: {kappa:.4f}")
from sklearn.metrics import confusion_matrix

# 计算混淆矩阵
conf_matrix = confusion_matrix(model.predict(test_data),test_labels)

# 计算制图精度（Producer's Accuracy）
producer_accuracy = np.diag(conf_matrix) / conf_matrix.sum(axis=0)

# 计算用户精度（User's Accuracy）
user_accuracy = np.diag(conf_matrix) / conf_matrix.sum(axis=1)

print("Producer's Accuracy:")
for i, accuracy in enumerate(producer_accuracy):
    print(f"Class {i}: {accuracy:.4f}")

print("\nUser's Accuracy:")
for i, accuracy in enumerate(user_accuracy):
    print(f"Class {i}: {accuracy:.4f}")
```
